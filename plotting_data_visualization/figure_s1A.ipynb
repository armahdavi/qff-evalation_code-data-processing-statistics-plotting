This figure compares the HVAC filter efficiency at the intial of and after being serviced in the studied home.

Plot type: Line plot

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from matplotlib import container
exec(open(r'C:\PhD Research\Generic Codes\notion_corrections.py').read())

df = pd.read_excel(backslash_correct('C:\PhD Research\QFF Evaluation\Processed\mini_wras_eff.xlsx'))
df = df[(df['Initial'] > 0) & (df['Final'] > 0) & (df['Overall'] > 0)]
```

```python
plt.figure()
plt.errorbar(df['Size'], df['Initial'], yerr = [df['Initial Error'], df['Initial Error']], xerr = None, 
             capsize = 2, c = "gray", ecolor = 'gray', elinewidth = 0.25, capthick = 0.25, label = 'Initial')

plt.errorbar(df['Size'], df['Final'], yerr = [df['Final Error'], df['Final Error']], xerr = None, 
             capsize = 2, c = "k", ecolor = 'black', elinewidth = 0.25, capthick = 0.25, label = 'Final')

plt.xscale('log')
   
a = [0.1, 0.3, 1, 3, 10, 30]
plt.xlim(0.1, 30)
plt.xticks(ticks = a, labels = a)
plt.minorticks_off()
plt.xlabel('Particle Size (Âµm)', size = 14)

plt.ylim(0, 100)
plt.yticks(np.arange(0, 101, step=20))
plt.ylabel('Efficiency (%)', size = 14)

plt.tick_params(axis = 'both', direction = 'inout')


## The below system doesn't work in loop as it cumulatvely add labels. Should later find a solution for it.
ax = plt.gca()
handles, labels = ax.get_legend_handles_labels()
new_handles = []
for h in handles:
    #only need to edit the errorbar legend entries
    if isinstance(h, container.ErrorbarContainer):
        new_handles.append(h[0])
    else:
        new_handles.append(h)
ax.legend(new_handles, labels, loc = 4, frameon = True, edgecolor = 'black')

plt.savefig(r'C:/PhD Research/QFF Evaluation/Processed/plots/Figure_s1a.jpg', format='jpg', dpi=1600, bbox_inches='tight')
```
